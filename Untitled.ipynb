```python
import google.generativeai as genai
import os
from dotenv import load_dotenv, find_dotenv
```

```python
load_dotenv(find_dotenv(), override=True)
api_key = os.environ.get('GOOGLE_API_KEY')
genai.configure(api_key=api_key)
```

# for prompts based on uploaded images

```python
#to take interactiveness and visuality of jupy-nb to next level
from IPython.display import Image, Markdown

# to display image
Image(filename='greece.jpg')
```

```python
# to upload file
img_file = genai.upload_file(path='greece.jpg', display_name="GREECE")
print(f'uploaded file:{img_file.display_name} as {img_file.uri}')

# to retrive the uploaded file
retrieved_img_file = genai.get_file(name=img_file.name)
print(f'uploaded file:{retrieved_img_file.display_name} as {retrieved_img_file.uri}')
```

```python
# to create model 
model = genai.GenerativeModel('gemini-1.5-pro-latest')

# to pass the image file, prompt and to get the response
prompt = 'describe the image considering historical facts'
response = model.generate_content([prompt, img_file])
print(response.text)
```
